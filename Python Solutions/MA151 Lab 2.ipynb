```python
from sympy import *
from sympy.plotting import (plot, plot_parametric)
```

## 1a volume of ice cream in cone

```python
L = 4
theta = 15*pi/180
r = L * sin(theta)
h = L * cos(theta)
Vol = 1/3*pi*r**2*h + 2/3*pi*r**3
print('The volume of ice cream is', Vol.evalf(),'cubic inches')
```

## 1b slant height needed for one cup

```python
L = symbols('L')
r = L * sin(theta)
h = L * cos(theta)
Vol = 1/3*pi*r**2*h + 2/3*pi*r**3
VolEqn = Vol - 14.5
soln = solve(VolEqn, L)
print('The slant height of the cone should be about', soln[0].evalf(), 'inches')
```

```python
t = symbols('t')
plot_parametric((7*cos(t),7*sin(t),(t,0,2*pi)),(3+5*cos(t),4+5*sin(t),(t,0,2*pi)))
```

## 3b intersections of circles

```python
t,s = symbols('t s')
x1 = 7.0*cos(t)
x2 = 3.0 + 5*cos(s)
y1 = 7.0*sin(t)
y2 = 4.0 + 5*sin(s)
slns = solve([x1-x2,y1-y2],[t,s])
print('The t and s solutions are', slns)
print('The first intersection is (',x1.subs(t,slns[0][0]),',', y1.subs(t,slns[0][0]),')')
print('The second intersection is (',x1.subs(t,slns[0][1]),',', y1.subs(t,slns[0][1]),')')
```
