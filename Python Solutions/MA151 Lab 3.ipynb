```python
from sympy import *
from sympy.plotting import (plot, plot_parametric)
```

## 1a Find the limit

```python
x = symbols('x')
g = 5 + x**2*cos(1/x**2)
L = limit(g,x,0)
print('The limit of g(x) as x approaches 0 is', L)
```

## 1b Illustrate Squeeze Theorem

```python
matplotlib notebook
```

```python
f = 5 - x**2
h = 5 + x**2
plot((f, (x,-0.5,0.5)), (g, (x,-0.5,0.5)), (h, (x,-0.5,0.5)))
```

## 2a Plot V(d)

```python
matplotlib notebook
```

```python
d = symbols('d')
M = 0.1
pW = 1000
C = 1
theta = 10*pi/180 #be sure to convert to radians
beta = 10*pi/180
g = 9.81
V = sqrt(16*M*g/(pi*C*pW*d**2))/sqrt(1-8*M*tan(beta)**2/(pi*d**3*C*pW*sin(theta)))
plot(V,(d,0.05,0.1),ylim=[0.7,0.9])
```

## 2b Theorem to find solution to V(d) = 0.8

```python
#answers can be within range
print('a = 0.09, b = 0.1')
print('By the Intermediate Value Theorem, there is a solution to V(d)=0.8 on [a,b]')
```

## 2c Numerical solution

```python
d_soln = nsolve(V-0.8,0.09)
print('V(d)=0.8 when d is approximately', d_soln)
```

## 3a List Comprehension 

```python
x = symbols('x')
f = sqrt(2*x**2+1)/(3*x-5)
to_inf = [f.subs(x,i) for i in [10.0,50.0,100.0]]
to_neginf = [f.subs(x,i) for i in [-10.0,-50.0,-100.0]]
print('y values to +oo are', to_inf)
print('y values to -oo are', to_neginf)
print('The limit to infinity appears to be about 0.47 and the limit to negative infinity appears to be about -0.47')
```

## 3b Compute Limits

```python
limit_inf = limit(f,x,oo)
print('The limit of f as x -> oo is', limit_inf, 'or approximately', limit_inf.evalf())
limit_neginf = limit(f,x,-oo)
print('The limit of f as x -> -oo is', limit_neginf, 'or approximately', limit_neginf.evalf())
```

## 3c Plot function and horizontal asymptotes

```python
matplotlib notebook
```

```python
plot((f,(x,-10,10)), (limit_inf,(x,-10,10)), (limit_neginf,(x,-10,10)), ylim = [-10,10])
```

## 4a Graph of piecewise function

```python
matplotlib notebook
```

```python
f1 = 2*x - x**2
f2 = 2-x
f3 = (x**2 - 7*x + 12)/(x-3)
f4 = pi
plot((f1,(x,0,2)), (f2,(x,2,3)), (f3,(x,3,4)), (f4,(x,4,6)))
print('f appears to be continuous at x=2 and x=3, but not x=4')
```

## 4b Left and Right Hand limits at 2, 3, 4

```python
limitleft2 = limit(f1,x,2)
print('The limit of f as x -> 2 from the left is', limitleft2)
limitright2 = limit(f2,x,2)
print('The limit of f as x -> 2 from the right is', limitright2)
print('f(2) =', f1.subs(x,2))
print('Therefore, f is continuous at x=2 since the limit exists and is equal to f(2)')
limitleft3 = limit(f2,x,3)
print('The limit of f as x -> 3 from the left is', limitleft3)
limitright3 = limit(f2,x,3)
print('The limit of f as x -> 3 from the left is', limitright3)
print('f(3) =', f2.subs(x,3))
print('Therefore, f is continuous at x=3 since the limit exists and is equal to f(3)')
limitleft4 = limit(f3,x,4)
print('The limit of f as x -> 4 from the left is', limitleft4)
limitright4 = limit(f4,x,4)
print('The limit of f as x -> 4 from the left is', limitright4)
print('f(4) =', f4.subs(x,4))
print('Therefore, f is continuous at x=4 since the limit exists and is equal to f(4)')
```
