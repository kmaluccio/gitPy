```python
from sympy import *
from sympy.plotting import (plot, plot_parametric)
```

## 1a Tangent line at x=2

```python
x = symbols('x')
f = 5 * x/(1+x**2)
df = diff(f,x)
mtan = df.subs(x,2)
yval = f.subs(x,2)
tanlineeqn = expand(mtan*(x-2)+yval) #expand makes the function look nicer
print('The equation of the tangent line is y =', tanlineeqn)
```

## 1b Graph of f and tangent line

```python
matplotlib notebook
```

```python
plot((f,(x,0,4)), (tanlineeqn,(x,0,4)))
```

## 2a Velocity after 1 second

```python
t = symbols('t', positive = True)
s = 1+10*t-1.86*t**2
vel = diff(s,t)
v1 = vel.subs(t,1)
print('The velocity after 1 second is', v1, 'm/s')
```

## 2b when v=0 and corresponding height

```python
vis0 = solve(vel,t)
smax = s.subs(t,vis0[0])
print('The velocity is 0 after', vis0[0], 'seconds')
print('Height at this time is', smax, 'meters')
```

## 2c velocity when hits the surface

```python
hitsurf = solve(s,t)[0]
vhit = vel.subs(t,hitsurf)
print('The rock will hit the surface with a velocity of', vhit, 'm/s')
```

## 3a plot of f

```python
x = symbols('x', real=True)
f = x*exp(-abs(x))
plot(f,(x,-5,5))
```

## 3b Equations of horizontal tangent lines

```python
x_neg = x*exp(x)
x_pos = x*exp(-x)
x_neg_crit_pt = solve(diff(x_neg,x), x)
x_pos_crit_pt = solve(diff(x_pos,x), x)
print('The equations are y =', f.subs(x,x_neg_crit_pt[0]), 'and y =', f.subs(x,x_pos_crit_pt[0]))
```

## 4a first 16 derivatives

```python
x = symbols('x')
f = x*sin(x)
derivs = [diff(f,(x,i)) for i in range(1,17)]
print('The first 16 derivatives are:', derivs)
```

## 4b formula for nth, (n+1)th, (n+2)th, (n+3)th derivative if n divisible by 4

```python
print("f^(n)(x) = x*sin(x) - n*cos(x)")
print("f^(n+1)(x) = x*cos(x) + (n+1)*sin(x)")
print("f^(n)(x) = -x*sin(x) + (n+2)*cos(x)")
print("f^(n)(x) = -x*cos(x) - (n+3)*sin(x)")
```
