```python
from sympy import *
from sympy.plotting import (plot, plot_parametric)
```

## 1a Values of r which solve differential equation

```python
r, x = symbols('r x')
f = E**(r*x)
df = diff(f,x)
ddf = diff(f,x,2)
deq = ddf + 5*df + 6*f
r_soln = solve(deq,r)
print('The values of r are', r_soln)
```

## 1b Values of r which solve differential equation

```python
deq2 = ddf + 6*df + 13*f
r_soln2 = solve(deq2, r)
print('The values of r are', r_soln2)
```

## 1c Compute left hand side with cosines and sines and conclusion

```python
y = E**(-3*x)*(cos(2*x)+sin(2*x))
dy = diff(y,x)
ddy = diff(y,x,2)
eqn = ddy + 6*dy + 13*y
print('The left hand side yields', eqn.simplify(), 'which means y satisfies the differential equation')
print('The exponential e raised to a complex exponent yields sines and cosines')
```

## 2a Implicit Plot

```python
matplotlib notebook
```

```python
x, y = symbols('x y', real = True) #want real values here, not imaginary
F = (x**2+y**2)**2 - 4*(x**2-y**2) + 3
Fplot = plot_implicit(F, (x,-3,3), (y,-3,3))
```

## 2b dy/dx

```python
dydx = idiff(F,y,x)
print('The implicit derivative of y is', dydx)
```

## 2c horizontal tangent lines

```python
htanpts = solve([F,dydx],x,y,real=True)
print('There are horizontal tangents at', htanpts)
```

## 2d Tangent line equation at (sqrt(2), sqrt(sqrt(17)-4))

```python
x1 = sqrt(2)
y1 = sqrt(sqrt(17)-4)
m = dydx.subs({x:x1,y:y1})
print(m)
tanline = m*(x-x1) + y1
print('The equation of the tangent line to the curve at the given point is y =', tanline)
```

## 2e Plot equation with tangent line from part d

```python
matplotlib notebook
```

```python
tangentplot = plot(tanline, (x,-3,3), ylim=[-3,3], show=False)
Fplot.extend(tangentplot)
Fplot.show()
```

## 3a (n+1)th derivative of x^n log(x)

```python
x, n = symbols('x n')
derivs = [diff(x**n*log(x), x, n+1) for n in range(2,9)]
print('For n = 2..8, the (n+1)th derivative of x^n*ln(x) is', derivs)
```

## 3b Conclusion

```python
print('In general, the (n+1)th derivative of x^n*ln(x) is n!/x')
```

## 4a logarithmic differentiation

```python
# in part a we take the log of both sides by hand and solve this way
x = symbols('x', real=True, positive=True)
y = x**(sqrt(x))
lny = sqrt(x)*log(x)
dlny = diff(lny,x)
dydx = y*dlny
print(dydx)
# can also use expand log command
lny = expand_log(log(y), force=True)
print(lny)
dlny = diff(lny,x)
dydx = y*dlny
print('Using logarithmic differentiation, the derivative is', dydx)
```
