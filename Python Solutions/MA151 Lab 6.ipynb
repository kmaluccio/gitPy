```python
from sympy import *
from sympy.plotting import (plot, plot_parametric)
```

# 1a Equation of tangent line when t=pi/4

```python
t = symbols('t')
x = cos(t)
y = sin(t)+3
x0 = x.subs(t,pi/4)
y0 = y.subs(t,pi/4)
dx = diff(x,t)
dy = diff(y,t)
dydx = dy/dx
m = dydx.subs(t,pi/4)
x = symbols('x')
tanline = m*(x-x0)+y0
print('The equation of the tangent line is y=', tanline, 'or approximately', tanline.evalf())
```

## 1b points where tangent line is vertical

```python
v_tan = solve(dx,t)
xt = cos(t)
print(v_tan)
print('The curve has vertical tangent lines at (', xt.subs(t,v_tan[0]),',',y.subs(t,v_tan[0]),') and at (', xt.subs(t,v_tan[1]), ',', y.subs(t,v_tan[1]),')')
```

## 1c Plot parametrized curve and all tangent lines

```python
matplotlib notebook
```

```python
plotcurve = plot_parametric(xt, y, (t,0,2*pi),xlim=[-4,4],ylim=[0,6], show=False)
plotv_tan1 = plot_parametric(1, t, (t,0,2*pi), show=False)
plotv_tan2 = plot_parametric(-1, t, (t,0,2*pi), show=False)
plottanpi = plot(tanline, (x,-4,4), ylim=[0,6], show=False)
plotcurve.extend(plotv_tan1)
plotcurve.extend(plotv_tan2)
plotcurve.extend(plottanpi)
plotcurve.show()
```

## 2a Tangent lines at (3,0)

```python
t, x = symbols('t x')
f = t**2
g = t**3 - 3*t
tsoln = solve([f-3,g],t)
print(tsoln)
df = diff(f,t)
dg = diff(g,t)
dydx = dg/df
m1 = dydx.subs(t,-sqrt(3))
m2 = dydx.subs(t,sqrt(3))
tanline1 = m1*(x-3)
tanline2 = m2*(x-3)
print('The equations of the tangent lines are y=', tanline1, 'and y=', tanline2)
```

## 2b Plot of parametrized curve and tangent lines

```python
matplotlib notebook
```

```python
pf = plot_parametric(f,g,(t,-2,2), show=False)
pf.extend(plot((-sqrt(3)*(x-3), (x,0,4)), (sqrt(3)*(x-3),(x,0,4)), show=False))
pf.show()
```

## 3a find k and y0

```python
k, y0 = symbols('k y0')
y = y0*E**(k*t)
eqn1 = y.subs(t,1) - 200
eqn2 = y.subs(t,Rational(3,2)) - 360
solns = solve([eqn1,eqn2], [k,y0])
print('k =', solns[0][0].evalf(), ' y0 =', solns[0][1].evalf())
```

## 3b When population = 2000

```python
y = solns[0][1]*E**(solns[0][0]*t)
tsoln = solve(y-2000,t)
print('Count reaches 2,000 after', tsoln[0],', or approximately', tsoln[0].evalf(),'hours')
```

## 3c find k and population 1 hour before "initial"

```python
newy = 200*E**(k*t)
newk = solve(newy.subs(t,0.5)-360,k)
y = 200*E**(newk[0]*t)
yatneg1 = y.subs(t,-1)
print('k = ', newk[0])
print('Count at t=-1 is', yatneg1, 'bacteria, which is the same as the original initial amount')
```

## 4a rate of change in f with respect to each variable

```python
L,T,p = symbols('L T p')
f = 1/(2*L)*sqrt(T/p)
dfdL = diff(f,L)
dfdT = diff(f,T)
dfdp = diff(f,p)
print('The derivative of f with respect to:')
print('L is', dfdL)
print('T is', dfdT)
print('rho is', dfdp)
```

## 4b interpret what happens to the pitch

```python
print('As the string length decreases, the frequency increases, creating a higher pitch.')
print('As the tension increases, the frequencey increases, creating a higher pitch.')
print('As the density increases, the frequency decreases, creating a lower pitch.')
```
