```python
from sympy import *
from sympy.plotting import (plot, plot_parametric)
```

## 1a Graph of piecewise function

```python
matplotlib notebook
```

```python
x = symbols('x')
f1 = 36 - x**2
f2 = 4*x - 2
plot((f1,(x,-6,0)), (f2, (x,0,6))) #Note we are not plotting the open/closed circles at x=0
```

## 1b Absolute max and min based on graph

```python
print('The absolute minimum of f(x) on [-6,6] is', f2.subs(x,0), 'which occurs at x=0')
print('The absolute maximum of f(x) on [-6,6] does not exist')
```

## 2a Critical values of f

```python
x = symbols('x', real=True)
f = E**x + E**(-3*x)
df = diff(f,x)
crit_vals = solve(df,x)
print('The critical value is', crit_vals, 'or approximately', crit_vals[0].evalf())
```

## 2b Absolute max and min on [0,1]

```python
xvals = [0, crit_vals[0], 1]
yvals = [f.subs(x,i).evalf() for i in xvals]
print('The y values are', yvals)
print('The absolute maximum is', yvals[2],'at x = 1')
print('The absolute minimum is', yvals[1],'at x = ',crit_vals[0])
```

## 2c Graphical confirmation of part b

```python
matplotlib notebook
```

```python
plot(f,(x,0,1))
```

## 3a Plot of f and apparent number of local extrema

```python
matplotlib notebook
```

```python
x = symbols('x')
f = sin(x+sin(3*x))
plot(f,(x,0,2*pi), ylim=[-1.2,1.2])
print('From the graph, it appears that f has 4 local maxima and 4 local minima on [0,2pi]')
```

## 3b Derivative

```python
df = diff(f,x)
print("f'(x)=",df)
```

## 3c Plot f' in [0.5,0.7]

```python
matplotlib notebook
```

```python
plot(df,(x,0.5,0.7))
```

## 3d Critical values in [0.5,0.7]

```python
c1 = nsolve(df,0.6)
c2 = nsolve(df,0.65)
c3 = nsolve(df,0.7)
print('The critical values between 0.5 and 0.7 are', c1,',', c2,'and', c3)
```

## 3e repeat parts c and d for the interval [5.5,5.7]

```python
matplotlib notebook
```

```python
plot(df,(x,5.5,5.7))
c4 = nsolve(df,5.6)
c5 = nsolve(df,5.65)
c6 = nsolve(df,5.7)
print('The critical values between 5.5 and 5.7 are', c4, ',',c5,'and',c6)
```

## 3f Actual number of local extrema

```python
print('There are 3 critical values between x=0.5 and x=0.7 and 3 critical values between x=5.5 and x=5.7')
print('Thus, there are a total of 12 critical values - 6 local maxima and 6 local minima')
```

## 4a Two variable function to be maximized and equation relating the two variables

```python
x, h, S = symbols('x h S')
V = x**2*h
Seqn = x**2 + 4*x*h
print('We wish to maximize V=',V)
print('The equation is',Seqn,'=S')
```

## 4b Solve for h and substitute into function

```python
solveforh = solve(Seqn-S,h)[0]
Vofx = V.subs(h,solveforh)
print('The new function is V(x)=',Vofx)
```

## 4c practical domain

```python
print('h=',solveforh) #note minimum is at h=0
xmax = solve(solveforh,x)
print(xmax) #positive value is the one we want
print('The practical domain of V(x) is (0,',xmax[1],']')
```

## 4d Absolute maximum of function from part b on domain from part c

```python
dV = diff(Vofx,x)
critvals = solve(dV,x)
print(critvals)
xval = [0, critvals[1],xmax[1]]
yval = [Vofx.subs(x,i) for i in xval]
print('The y values are', yval)
print('The maximum volume is',yval[1], 'when x =',critvals[1])
print('and h=', solveforh.subs(x,critvals[1]))
```
