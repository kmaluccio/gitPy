```python
from sympy import *
from sympy.plotting import (plot, plot_parametric)
```

## 1a find and simplify the derivative of f-g

```python
x = symbols('x', positive=True)
f = 2*asin(x)
g = acos(1-2*x**2)
dfminusg = diff(f-g,x)
print('The derivative of f(x)-g(x) =', dfminusg.simplify())
```

## 1b, 1c Implication of f-g and specific answer

```python
print('This tells us that f(x)-g(x) is a constant')
print(f.subs(x,0)-g.subs(x,0),'is the value of the constant, so f(x)=g(x)')
```

## 2a critical values of g

```python
x = symbols('x')
g = -5*x**4 - 20*x**3 + 18
dg = diff(g,x)
critvals = solve(dg,x)
print('The critical values of the function g(x) are', critvals)
```

## 2b intervals where g is increasing/decreasing

```python
matplotlib notebook
```

```python
plot(dg,(x,-4,1)) #checked domain here and this works nicely
print('From the graph, g is increasing on (-oo,-3)')
print('and decreasing on (-3,0) U (0,oo)')
```

## 2c where g'' = 0

```python
ddg = diff(g,x,2)
ivalues = solve(ddg,x)
print(ivalues)
```

## 2d intervals of concavity of g

```python
matplotlib notebook
```

```python
plot(ddg,(x,-3,1))
print('From the graph, g is concave up on (-2,0)')
print( 'and concave down on (-oo,-2) U (0,oo)')
```

## 3a plot of f and number of extrema and inflection points based on graph

```python
matplotlib notebook
```

```python
x = symbols('x',real=True)
f = (x**2*(x+1)**3)/((x-2)**2*(x-4)**4)
plot(f,(x,-10,10))
print('There appear to be no local extrema or inflection points')
```

## 3b f' and critical values

```python
df = diff(f,x)
print('The derivative of f is', df.simplify())
cvals = solve(df,x)
print(cvals)
cvalsdec = [i.evalf() for i in cvals]
print('The critical values are', cvalsdec)
```

## 3c intervals of increase or decrease (attempted both)

```python
matplotlib notebook
```

```python
plot(df,(x,-21,5),ylim=[-1,1])
testvalues = [-21, -5, -0,5, -0.2, 1, 2.2, 3, 5]
dfsign = [df.subs({x:i}) for i in testvalues]
print('The y values of the first derivative at each test point are', dfsign)
print('So, f is increasing on (-20.14,-1) U (-1,-0.32) U (0,2) U (2.47,4)')
print('and decreasing on (-oo,-20.14) U (-0.32,0) U (2,2.47) U (4,oo)')
```

## 3d intervals of concavity

```python
ddf = diff(df, x)
print('The second derivative is', ddf.factor())
c2values = solve(ddf,x)
c2values_float = [i.evalf() for i in c2values]
print('The inflection points are', c2values_float)
testvals = [-40,-10,-2,-0.7,-0.3,0,3,5]
ddfsign = [ddf.subs({x:i}) for i in testvals]
print('The y values of the second derivative at each test point are', ddfsign)
print('So f is concave up on (-35.31, -4.98) U (-1,-0.54) U (-0.11,2) U (2,4) U (4,oo)')
print('and concave down on (-oo,-35.31) U (-4.98,-1) U (-0.54,-0.11)')
```

## 4a Define f and g for natural log of y

```python
x = symbols('x',real=True)
y = (1-6*x)**(1/x)
# take log of both sides and write f and g
f = ln(1-6*x)
g = x
print('ln(y)=f(x)/g(x), where f=',f,'and g=',g)
```

## 4b limits of f and g as x approaches 0

```python
F = limit(f,x,0)
G = limit(g,x,0)
print('f(x) approaches', F, 'and g(x) approaches',G,'as x approaches 0')
```

## 4c Apply L'Hospital's Rule (if applicable)

```python
print("Since conditions for l'hospitals rule are met, we can apply it here")
df = diff(f,x)
dg = diff(g,x)
newF = limit(df,x,0)
newG = limit(dg,x,0)
print("f'(x) approaches", newF,"and g'(x) approaches", newG,'as x approaches 0')
print('Therefore, ln(y) approaches',newF/newG,'and y therefore approaches', exp(newF/newG))
```
